```python
import pandas as pd
import matplotlib.pyplot as mp
```

```python
grouped = a.groupby("Description")["Quantity"].sum().sort_values(ascending = False).head(10)
mp.figure(figsize=(10,6))
mp.bar(grouped.index, grouped)
mp.xlabel("Total Quantity Sold")
mp.ylabel("Product")
mp.title("Top 10 High-Demand Products")
mp.xticks(rotation=30)
mp.show()
```

```python
grouped2 = a.groupby("Description")["Quantity"].sum().sort_values().head(10)
mp.figure(figsize=(10,6))
mp.bar(grouped2.index, grouped2)
mp.xlabel("Total Quantity Sold")
mp.ylabel("Product")
mp.title("Top 10 Low-Demand Products")
mp.xticks(rotation=45)
mp.gca().invert_yaxis()
mp.show()
```

```python
grouped3 = a.groupby("Country")["Quantity"].sum().sort_values(ascending = False).head(5)

mp.figure(figsize=(10,6))
mp.plot(grouped3.index, grouped3)
mp.xlabel("Total Quantity Sold")
mp.ylabel("Product")
mp.title("Top 5 Countries by Products")
mp.xticks(rotation=30)

mp.show()
```
